```python
class representation:
    def __init__(self,W):
        self.W = W
        self.n = [w.shape[1] for w in W] + [W[-1].shape[0]]
        self.L = len(self.n)-1
        self.flag = False
        
    def padzeros(self,M,newrows,newcols = None):
        oldrows, oldcols = M.shape
        if newcols == None:
            newcols = oldcols
        return np.pad(M,((0,newrows-oldrows),(0,newcols-oldcols)),mode="constant")
    
    def decomposeWeights(self):
        n = self.n
        
        g = []
        U = [np.zeros((n[1],n[0]))]
        W_red = []
        W_current = self.W[0]

        for s in range(self.L-1):
            Q,R = np.linalg.qr(W_current, mode="complete")
            g.append(Q)
            W_red.append(R[:n[0]])  
            QP = np.copy(Q)
            QP[:,:n[0]] = np.zeros((n[s+1],n[0]))
            U.append(self.W[s+1] @ QP @ np.transpose(Q) )
            W_current = (self.W[s+1] @ Q)[:,:n[0]]

        W_red.append(W_current)
        
        self.g = g
        self.U = U
        self.W_red = W_red
        self.flag = True
        
        return (W_red,U,g)

    def testDecomp(self):
        if not self.flag:
            self.decomposeWeights()
        
        #For Formula Readability.  
        U = self.U
        W_red = self.W_red
        g = self.g
        L = self.L 
        
        n = self.n
        
        #Reconstructed W
        W_test = ([U[0] + g[0] @ self.padzeros(W_red[0],n[1])] + 
          [U[s] + g[s] @ self.padzeros(W_red[s],n[s+1],n[s]) @ np.transpose(g[s-1]) for s in range(1,L-1)] + 
          [U[L-1] + self.padzeros(W_red[L-1],n[L],n[L-1]) @ np.transpose(g[L-2])])
        
        return (max([np.max(np.abs(w1-w2)) for w1,w2 in zip(self.W,W_test)]) < 1e-10)
    
    def printShapes(self):
        print("Wred shapes:",[w.shape for w in self.W_red])
        print("U    shapes:",[u.shape for u in self.U])
        print("G    shapes:",[x.shape for x in self.g])
```

```python
def testRepresentation(verbose=False):
    n = [4, 64, 16, 128]
    L = len(n)-1
    W = [np.random.rand(n[i+1],n[i]) for i in range(L)]
    if verbose:
        for w in W:
            print(w.shape)
    rep = representation(W)
    rep.decomposeWeights()
    return rep.testDecomp()
```

```python
all([testRepresentation() for s in range(100)])
```
